```python
#222002111, Md Syful Islam, CSE 412, 222-D3.

import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python

df = pd.read_csv('/content/winequality-red.csv')

missing_values = df.isnull().sum()

print("Missing values per column:")
print(missing_values)


if missing_values.sum() > 0:
    print("\nMissing values are present. Handling methods")
    print("-Filling missing values with a calculated value .")
    print("- Dropping rows or columns: Removing rows or columns that contain missing values.")
    print("The specific method used depends on the nature of the data and the problem.")
else:
    print("\nNo missing values found in the dataset.")
```

```python
numerical_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar']
df_numerical = df[numerical_cols]


scaler_standard = StandardScaler()
df_standardized = scaler_standard.fit_transform(df_numerical)
df_standardized = pd.DataFrame(df_standardized, columns=numerical_cols)

print("Standardized Data:")
display(df_standardized.head())

scaler_minmax = MinMaxScaler()
df_normalized = scaler_minmax.fit_transform(df_numerical)
df_normalized = pd.DataFrame(df_normalized, columns=numerical_cols)

print("\nNormalized Data (Min-Max Scaling):")
display(df_normalized.head())
```

```python
low_quality_threshold = 4
high_quality_threshold = 7

def categorize_quality(quality):
    if quality <= low_quality_threshold:
        return 'low'
    elif quality >= high_quality_threshold:
        return 'high'
    else:
        return 'medium'

df['quality_group'] = df['quality'].apply(categorize_quality)

print("Distribution of Quality Groups:")
print(df['quality_group'].value_counts())

print("\nDataFrame with 'quality_group' column:")
display(df.head())
```

```python
X = df.drop(['quality', 'quality_group'], axis=1)
y = df['quality_group']

print("Features (X):")
display(X.head())

print("\nTarget variable (y):")
display(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
nb_model = GaussianNB()

nb_model.fit(X_train, y_train)

print("Naive Bayes model trained successfully.")
```

```python
y_pred = nb_model.predict(X_test)

print("First few predictions:")
print(y_pred[:5])
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)
```
